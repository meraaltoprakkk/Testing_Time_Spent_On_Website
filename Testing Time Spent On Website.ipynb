```python
# Problem : Web sitemizde geçirilen ortalama süre gerçekten 170 saniye mi?
# Detaylar : 
 # Yazılımlardan elde edilen web sitesinde geçirilen ort. süreler var.
 # Bu veriler incelendiğinde bir yönetici yada çalışanımız bu değerlerin 
 # böyle olmadığına yönelik düşünceler taşıyor ve bu durumu test etmek istiyorlar.
```

```python
import numpy as np

olcumler = np.array([17, 160, 234, 149, 145, 107, 197, 75, 201, 225, 211, 119,
                     157, 145, 127, 244, 163, 114, 145, 65, 112, 185, 202, 146,
                     203, 224, 203, 114, 188, 156, 187, 154, 117, 95, 165, 50, 110,
                     216, 138, 151, 166, 135, 155, 84, 251, 173, 131, 207, 121, 120])
```

```python
#Web sitemizde geçirilen zamanlar
olcumler[0:10]
```

```python
#Ölçümlerin Betimsel İstatistikleri
import scipy.stats as stats
```

```python
stats.describe(olcumler)
```

```python
#Varsayımlar:
#Normallik Varsayımı
```

```python
import pandas as pd
```

```python
#Histogram Tekniği
pd.DataFrame(olcumler).plot.hist();
```

```python
#qqplot 
import pylab
stats.probplot(olcumler, dist = "norm", plot = pylab)
pylab.show()
```

```python
#Shapiro- Wilk Testi
```

```python
#Ho : Örnek dağıımı ile teorik normal dağılım arasında istatistiksel olarak 
#anlamlı bir fark yokturç
#H1 : ..... fark vardır..
```

```python
from scipy.stats import shapiro
```

```python
shapiro(olcumler)
```

```python
print("T Hesap İstatistiği : " + str(shapiro(olcumler)[0]))
print("Hesaplanan P-value: " + str(shapiro(olcumler)[1]))
```

```python
#Hipotez Testinin Uygulanması :
```

```python
stats.ttest_1samp(olcumler, popmean = 170)
```

```python
#HO: Web sitemizde geçirilen ortalama süre 170'tir.
#H1: ......... değişir.
#H0<0.05 olduğundan dolayı ; H0 Reddedilir.
# Dolayısıyla ; Web Sitemizde geçirilen ortalam süremiz 170 değildir.
```
